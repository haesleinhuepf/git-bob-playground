# Christmas Tree Bauble Analysis
First, let's import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Load and display the image

```python
img = plt.imread('christmas_tree.png')
plt.imshow(img)
plt.axis('off')
plt.savefig('original.png', dpi=150, bbox_inches='tight')
```

## Analyze red baubles

```python
red_mask = (img[:,:,0] > 200) & (img[:,:,1] < 100) & (img[:,:,2] < 100)
plt.imshow(red_mask, cmap='gray')
plt.title('Red Baubles')
plt.axis('off')
plt.savefig('red_baubles.png', dpi=150, bbox_inches='tight')
```

## Analyze blue baubles

```python
blue_mask = (img[:,:,2] > 200) & (img[:,:,1] < 100) & (img[:,:,0] < 100)
plt.imshow(blue_mask, cmap='gray')
plt.title('Blue Baubles')
plt.axis('off')
plt.savefig('blue_baubles.png', dpi=150, bbox_inches='tight')
```

## Count baubles using connected components

```python
from scipy.ndimage import label
red_count, _ = label(red_mask)
blue_count, _ = label(blue_mask)

results = {
    'red': np.max(red_count),
    'blue': np.max(blue_count)
}

with open('bauble_counts.txt', 'w') as f:
    f.write(f"Red baubles: {results['red']}\n")
    f.write(f"Blue baubles: {results['blue']}\n")
```
