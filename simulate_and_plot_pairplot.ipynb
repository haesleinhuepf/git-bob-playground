# Simulate Data and Plot Pairplot

In this notebook, we will simulate a dataset of 100 patients and generate three different measurements for each patient. We will also include a classification stage for each patient, which can be either 1, 2, or 3. Finally, we will use Seaborn to create a pairplot of these measurements using the stage classification as hue.

## Step 1: Import Required Libraries

We will start by importing necessary libraries: `numpy`, `scipy`, `pandas`, and `seaborn`.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2: Simulate the Dataset

We'll simulate data for 100 patients with three measurements. Each measurement will be drawn from a normal distribution with different means and standard deviations. Additionally, we will randomly assign each patient a stage classification (1, 2, or 3).

```python
np.random.seed(0)  # For reproducibility
n_patients = 100

# Simulate measurements
measurement_1 = np.random.normal(loc=50, scale=10, size=n_patients)
measurement_2 = np.random.normal(loc=60, scale=15, size=n_patients)
measurement_3 = np.random.normal(loc=70, scale=20, size=n_patients)

# Simulate stage classifications
stage = np.random.choice([1, 2, 3], size=n_patients)

# Create DataFrame
data = pd.DataFrame({
    'Measurement 1': measurement_1,
    'Measurement 2': measurement_2,
    'Measurement 3': measurement_3,
    'Stage': stage
})

data.head()
```

## Step 3: Plot the Pairplot

We'll now use Seaborn's `pairplot` function to create a pairplot of the measurements and use the 'Stage' column as the hue.

```python
# Plot pairplot
sns.pairplot(data, hue='Stage')

# Save the pairplot to a file
plt.savefig('pairplot.png')
plt.show()
```
