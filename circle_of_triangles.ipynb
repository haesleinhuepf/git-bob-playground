# Generating a Circle of Triangles

In this notebook, we'll use NumPy and Seaborn to generate and visualize a dataset of 2D points that form 100 small triangles arranged in a large circle.

```python
# Import necessary libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for seaborn
sns.set(style="whitegrid")
```

## Generate 100 small triangles arranged in a big circle

We'll generate points such that they form small triangles arranged in a large circle.

```python
# Parameters for the big circle
num_triangles = 100
big_circle_radius = 10
small_triangle_size = 0.5

# Generate angles for the placement of triangles
angles = np.linspace(0, 2 * np.pi, num=num_triangles, endpoint=False)

# Function to create a single triangle at a given angle
def create_triangle(angle, radius, size):
    center_x = radius * np.cos(angle)
    center_y = radius * np.sin(angle)
    
    # Calculate vertices of the triangle
    vertices = np.array([
        [center_x + size * np.cos(angle), center_y + size * np.sin(angle)],
        [center_x + size * np.cos(angle + 2*np.pi/3), center_y + size * np.sin(angle + 2*np.pi/3)],
        [center_x + size * np.cos(angle - 2*np.pi/3), center_y + size * np.sin(angle - 2*np.pi/3)]
    ])
    return vertices

# Create all triangles
triangles = [create_triangle(angle, big_circle_radius, small_triangle_size) for angle in angles]
```

## Visualize the triangles
Using matplotlib and seaborn for visualization.

```python
# Plot the triangles
plt.figure(figsize=(8, 8))

for triangle in triangles:
    triangle = np.append(triangle, [triangle[0]], axis=0)  # Close the triangle
    plt.plot(triangle[:, 0], triangle[:, 1], "b-")

plt.title("Circle of Triangles")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")

# Save plot as PNG
plt.savefig("circle_triangles.png")
plt.show()
```
