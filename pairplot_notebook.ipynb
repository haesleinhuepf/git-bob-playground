# Pair-Plot of Simulated Patient Data

In this notebook, we'll simulate a dataset of 100 patients with three measurements and a stage classification. We'll then create a pair-plot to visualize these measurements with their stage classifications as the hue.

## Step 1: Import Libraries

Let's begin by importing the necessary libraries: pandas for data manipulation, numpy for numerical operations, and seaborn for visualization.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2: Simulate the Dataset

We'll create a DataFrame with 100 patients, with three random measurements, and randomly assigned stages (1, 2, or 3).

```python
# Seed for reproducibility
np.random.seed(0)

# Simulate measurements
measurements = np.random.rand(100, 3) * 100  # Example range from 0 to 100

# Simulate stage classification
stages = np.random.choice([1, 2, 3], size=100)

# Create DataFrame
df = pd.DataFrame(measurements, columns=['Measurement1', 'Measurement2', 'Measurement3'])
df['Stage'] = stages

# Display the first few rows of the DataFrame
df.head()
```

## Step 3: Create the Pair-Plot

Now, we'll create a pair-plot using Seaborn, setting the `hue` parameter to the stage classification.

```python
# Create the pairplot
pairplot = sns.pairplot(df, hue='Stage')

# Save the pairplot
pairplot.savefig('pairplot.png')

# Show the plot
plt.show()
```
