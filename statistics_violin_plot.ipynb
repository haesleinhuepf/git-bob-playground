# Violin Plot of Combined Normal Distributions

In this notebook, we will generate two 1D numeric datasets, each being a combination of three normal distributions with different means and standard deviations. We will then use these datasets to draw a violin plot.

## Step 1: Import Libraries

We will use `numpy` to create the datasets, `scipy.stats` for normal distribution, and `seaborn` for visualization.

```python
import numpy as np
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2: Generate Example Datasets

We will generate two datasets, `data1` and `data2`, each as a combination of three normal distributions with specified means and standard deviations.

```python
# Dataset 1
mean1_1, std1_1, n_samples1_1 = 0, 1, 500
mean1_2, std1_2, n_samples1_2 = 3, 0.5, 500
mean1_3, std1_3, n_samples1_3 = -3, 0.8, 500

data1_1 = norm.rvs(mean1_1, std1_1, size=n_samples1_1)
data1_2 = norm.rvs(mean1_2, std1_2, size=n_samples1_2)
data1_3 = norm.rvs(mean1_3, std1_3, size=n_samples1_3)

data1 = np.concatenate([data1_1, data1_2, data1_3])
```

```python
# Dataset 2
mean2_1, std2_1, n_samples2_1 = 2, 1.5, 500
mean2_2, std2_2, n_samples2_2 = -1, 0.5, 500
mean2_3, std2_3, n_samples2_3 = 4, 0.8, 500

data2_1 = norm.rvs(mean2_1, std2_1, size=n_samples2_1)
data2_2 = norm.rvs(mean2_2, std2_2, size=n_samples2_2)
data2_3 = norm.rvs(mean2_3, std2_3, size=n_samples2_3)

data2 = np.concatenate([data2_1, data2_2, data2_3])
```

## Step 3: Create Violin Plot

Finally, we will create a violin plot to compare the distributions of `data1` and `data2`.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=[data1, data2])
plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
plt.title('Violin Plot of Combined Normal Distributions')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.savefig('violin_plot.png')
plt.show()
```
