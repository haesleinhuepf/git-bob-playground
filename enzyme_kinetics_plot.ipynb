# Enzyme Kinetics Plot
Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Define functions for Michaelis-Menten and Hill kinetics

```python
def michaelis_menten(S, Km):
    return S / (Km + S)

def hill_equation(S, Km, n):
    return S**n / (Km**n + S**n)
```

Create data and plot

```python
# Generate substrate concentrations
S = np.logspace(-2, 2, 1000)

# Calculate velocities
Km = 1.0
v_mm = michaelis_menten(S, Km)
v_hill = hill_equation(S, Km, n=4)

# Create plot
plt.figure(figsize=(8, 6))
plt.plot(S, v_mm, 'r-', label='Non-cooperative')
plt.plot(S, v_hill, 'b-', label='Cooperative')
plt.axvline(x=Km, color='gray', linestyle='--', alpha=0.5)
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5)

plt.xscale('log')
plt.xlabel('[S]')
plt.ylabel('v')
plt.legend()
plt.grid(True)

# Save and show plot
plt.savefig('enzyme_kinetics.png')
plt.show()
```
