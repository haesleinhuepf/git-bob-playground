Import necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

Create the green channel (axons)

```python
green = np.zeros((256, 256, 3), dtype=np.uint8)
# Draw some lines for axons
for _ in range(20):
    x1, y1 = np.random.randint(0, 256, size=2)
    x2, y2 = np.random.randint(0, 256, size=2)
    plt.plot([x1, x2], [y1, y2], color='green', linewidth=np.random.rand()*2 + 1)
```

Create the red channel (cell bodies/synapses)

```python
for _ in range(50):
    x, y = np.random.randint(0, 256, size=2)
    plt.scatter(x, y, color='red', s=np.random.rand()*10 + 5)
```

Create the blue channel (nuclei)

```python
for _ in range(100):
    x, y = np.random.randint(0, 256, size=2)
    plt.scatter(x, y, color='blue', s=np.random.rand()*5 + 2)
```

Save and show the plot

```python
plt.axis('off')
plt.savefig("redraw_histology.png")
plt.show()
```
