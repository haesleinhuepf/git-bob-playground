## Segmentation Simulation

This notebook simulates an image, applies a Gaussian blur, normalizes the image, applies a threshold, performs connected component labeling, and counts the number of objects.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from skimage.measure import label, regionprops
```

### Create a 100x100 image with 10 random pixels set to 1

```python
image = np.zeros((100, 100))
random_pixels = np.random.randint(0, 100, size=(10, 2))
for pixel in random_pixels:
    image[pixel[0], pixel[1]] = 1
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.savefig('original_image.png')
plt.show()
```

### Apply Gaussian blur

```python
blurred_image = gaussian_filter(image, sigma=1)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
plt.savefig('blurred_image.png')
plt.show()
```

### Normalize the image to the range 0-1

```python
normalized_image = blurred_image / blurred_image.max()
plt.imshow(normalized_image, cmap='gray')
plt.title('Normalized Image')
plt.savefig('normalized_image.png')
plt.show()
```

### Apply a threshold of 0.5

```python
thresholded_image = normalized_image > 0.5
plt.imshow(thresholded_image, cmap='gray')
plt.title('Thresholded Image')
plt.savefig('thresholded_image.png')
plt.show()
```

### Perform connected component labeling

```python
labeled_image = label(thresholded_image)
plt.imshow(labeled_image, cmap='nipy_spectral')
plt.title('Labeled Image')
plt.colorbar()  # Add a color bar to show the color scale
plt.savefig('labeled_image.png')
plt.show()
```

### Count the number of objects

```python
num_objects = len(regionprops(labeled_image))
print(f'Number of objects: {num_objects}')
```
