# Drawing a cat using matplotlib

This notebook draws a detailed cat using matplotlib drawing functions.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Draw cat face
circle = plt.Circle((0.5, 0.5), 0.3, color='gray', fill=True)
ax.add_patch(circle)

# Draw ears (triangles)
left_ear = plt.Polygon([(0.3, 0.7), (0.4, 0.9), (0.5, 0.7)], color='gray')
right_ear = plt.Polygon([(0.5, 0.7), (0.6, 0.9), (0.7, 0.7)], color='gray')
ax.add_patch(left_ear)
ax.add_patch(right_ear)

# Draw eyes
left_eye = plt.Circle((0.4, 0.55), 0.05, color='green')
right_eye = plt.Circle((0.6, 0.55), 0.05, color='green')
left_pupil = plt.Circle((0.4, 0.55), 0.02, color='black')
right_pupil = plt.Circle((0.6, 0.55), 0.02, color='black')
ax.add_patch(left_eye)
ax.add_patch(right_eye)
ax.add_patch(left_pupil)
ax.add_patch(right_pupil)

# Draw nose
nose = plt.Circle((0.5, 0.45), 0.03, color='pink')
ax.add_patch(nose)

# Draw mouth
plt.plot([0.45, 0.5, 0.55], [0.4, 0.35, 0.4], 'k-', linewidth=2)

# Draw whiskers
whisker_starts = [(0.4, 0.45), (0.4, 0.42), (0.4, 0.39),
                  (0.6, 0.45), (0.6, 0.42), (0.6, 0.39)]
whisker_ends = [(0.2, 0.5), (0.2, 0.42), (0.2, 0.34),
                (0.8, 0.5), (0.8, 0.42), (0.8, 0.34)]
for start, end in zip(whisker_starts, whisker_ends):
    plt.plot([start[0], end[0]], [start[1], end[1]], 'k-', linewidth=1)

# Set axis properties
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

# Save and show the plot
plt.savefig('drawn_cat.png', dpi=300, bbox_inches='tight')
plt.show()
```
