# Drawing a cat using matplotlib
Let's draw a detailed cat portrait using matplotlib's drawing capabilities.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a new figure
plt.figure(figsize=(10, 8))

# Draw cat's head (main circular shape)
circle = plt.Circle((0.5, 0.5), 0.3, color='gray', fill=True)
plt.gca().add_patch(circle)

# Draw triangular ears
plt.plot([0.3, 0.4, 0.5], [0.7, 0.9, 0.7], 'gray', fill=True)
plt.plot([0.5, 0.6, 0.7], [0.7, 0.9, 0.7], 'gray', fill=True)

# Draw eyes (almonds)
plt.gca().add_patch(plt.Ellipse((0.4, 0.55), 0.1, 0.05, angle=-15, color='green'))
plt.gca().add_patch(plt.Ellipse((0.6, 0.55), 0.1, 0.05, angle=15, color='green'))

# Draw nose (triangle)
plt.plot([0.48, 0.5, 0.52], [0.48, 0.45, 0.48], 'k-')

# Draw whiskers
for x, y in [(0.3, 0.45), (0.7, 0.45)]:
    for angle in [-15, 0, 15]:
        dx = 0.15 * np.cos(np.radians(angle))
        dy = 0.15 * np.sin(np.radians(angle))
        plt.plot([x, x + dx], [y, y + dy], 'k-', linewidth=0.5)

# Draw mouth (curved line)
t = np.linspace(0.45, 0.55, 100)
plt.plot(t, 0.42 + 0.02 * np.sin((t-0.5) * 20), 'k-')

# Set axis limits and remove axes
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.axis('off')

# Save the plot
plt.savefig('drawn_cat.png', bbox_inches='tight', dpi=300)
plt.show()
```
