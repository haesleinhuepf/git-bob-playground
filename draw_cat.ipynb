# Drawing a cat with matplotlib
This notebook draws a cat using matplotlib shapes and curves.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure
fig, ax = plt.subplots(figsize=(8, 8))

# Head (circle)
circle = plt.Circle((0.5, 0.5), 0.3, color='gray', fill=True)
ax.add_patch(circle)

# Ears (triangles)
left_ear = plt.Polygon([(0.3, 0.7), (0.4, 0.9), (0.5, 0.7)], color='gray')
right_ear = plt.Polygon([(0.5, 0.7), (0.6, 0.9), (0.7, 0.7)], color='gray')
ax.add_patch(left_ear)
ax.add_patch(right_ear)

# Eyes (circles)
left_eye = plt.Circle((0.4, 0.55), 0.05, color='white')
right_eye = plt.Circle((0.6, 0.55), 0.05, color='white')
ax.add_patch(left_eye)
ax.add_patch(right_eye)

# Pupils (circles)
left_pupil = plt.Circle((0.4, 0.55), 0.02, color='black')
right_pupil = plt.Circle((0.6, 0.55), 0.02, color='black')
ax.add_patch(left_pupil)
ax.add_patch(right_pupil)

# Nose (triangle)
nose = plt.Polygon([(0.48, 0.45), (0.52, 0.45), (0.5, 0.42)], color='pink')
ax.add_patch(nose)

# Whiskers (lines)
whisker_starts = [(0.4, 0.45), (0.4, 0.43), (0.4, 0.41),
                  (0.6, 0.45), (0.6, 0.43), (0.6, 0.41)]
whisker_ends = [(0.2, 0.5), (0.2, 0.43), (0.2, 0.36),
                (0.8, 0.5), (0.8, 0.43), (0.8, 0.36)]
for start, end in zip(whisker_starts, whisker_ends):
    plt.plot([start[0], end[0]], [start[1], end[1]], 'k-', linewidth=1)

# Set axis limits and aspect ratio
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

# Save and show plot
plt.savefig('drawn_cat.png', dpi=300, bbox_inches='tight')
plt.show()
```
