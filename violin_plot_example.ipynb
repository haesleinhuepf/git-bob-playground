# Violin Plot of Mixed Normal Distributions

This notebook demonstrates how to generate two datasets composed of mixtures of three normal distributions and visualize them using a violin plot.

## Step 1: Import Required Libraries

We will use the following Python libraries:
- `numpy` for numerical operations.
- `scipy.stats` to generate normal distributions.
- `seaborn` for creating the violin plot.
- `matplotlib.pyplot` for displaying the plot.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## Step 2: Define a Function to Generate Mixed Normal Data

This function generates a dataset composed of multiple normal distributions with specified means and standard deviations.

```python
def generate_mixed_normal_data(size, means, std_devs):
    data = []
    num_distributions = len(means)
    size_per_distribution = size // num_distributions
    
    for mean, std in zip(means, std_devs):
        data.append(norm.rvs(loc=mean, scale=std, size=size_per_distribution))
    
    return np.concatenate(data)
```

## Step 3: Generate Example Datasets

We will generate two datasets, each a mixture of three normal distributions with different means and standard deviations.

```python
data1 = generate_mixed_normal_data(300, [0, 5, 10], [1, 2, 1.5])
data2 = generate_mixed_normal_data(300, [2, 7, 12], [2, 1.5, 1])
```

## Step 4: Visualize with a Violin Plot

Use the `seaborn` library to create and display a violin plot of the datasets.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=[data1, data2])
plt.title('Violin Plot of Mixed Normal Distributions')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.savefig('violin_plot.png')
plt.show()
```
